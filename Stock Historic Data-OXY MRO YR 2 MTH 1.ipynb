```python
mylist = [] # creating an empty list
for list_item in mylist:
    print(list_item)
```

```python
for x in range(len(df_concat.index)):
    var_df = df_concat['close_y'].iloc[x] - df_concat['close_x'].iloc[x]
    print(df_concat.index[x])
    print(var_df)
    mylist.append(var_df)

    
    

#for x in range(len(df_concat.index)):
#   print(df_concat['time'].iloc[x])
#   print(df_concat['close_x'].iloc[x])
#   print(df_concat['close_y'].iloc[x])
#   print(x
```

```python
mylist
```

```python
def Average(lst):
    return sum(lst) / len(lst)
```

```python
average = Average(mylist)
 
# Printing average of the list
print("Average of the list =", round(average, 2))
```

```python
# Standard deviation of list
# Using sum() + list comprehension
mean = sum(mylist) / len(mylist)
variance = sum([((x - mean) ** 2) for x in mylist]) / len(mylist)
res = variance ** 0.5
 
# Printing result
print("Standard deviation of sample is : " + str(res))
```
